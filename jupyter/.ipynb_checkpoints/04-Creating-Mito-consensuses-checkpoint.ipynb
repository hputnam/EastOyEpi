# Step 1 Separate Samples from VCF

```python
%%bash
for file in /Users/sr320/Documents/GitHub/EastOyEpi/data/SNP.TRSdp5g75mtDNA*vcf.gz; do
  for sample in `/Applications/bioinfo/bcftools-1.9/bcftools query -l $file`; do
    /Applications/bioinfo/bcftools-1.9/bcftools view -c1 -Oz -s $sample -o ${file/.vcf*/.$sample.vcf.gz} $file
  done
done
```

# Step 2 Extracting Consensus for each sample

```python
%%bash
find ../data/SNP.TRSdp5g75mtDNA.recode.*.vcf.gz \
| xargs basename -s .vcf.gz | xargs -I{} cat \
/Users/sr320/Desktop/oakl/genome/GCF_002022765.2_C_virginica-3.0_genomic.fa \
| /Applications/bioinfo/bcftools-1.9/bcftools consensus ../data/{}vcf.gz \
> ../analyses/{}consensus.fa
```
